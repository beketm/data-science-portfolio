```python
import pyforest
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
np.random.seed(22)
```

```python
#loading dataset
X, y = load_digits(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=22)
```

```python
nn_input = X_train.shape[1]
nn_output = 10
num_examples = X_train.shape[0]
nn_nodes = 10
learn_rate = 0.1
```

```python
def forward(model, data_x):
    W1, b1, W2, b2 = model["W1"], model["b1"], model["W2"], model["b2"] # Weights and biases 
    
    z1 = data_x.dot(W1) + b1
    a1 = np.tanh(z1)
    
    z2 = a1.dot(W2) + b2
    exp_vals = np.exp(z2)
    probs = exp_vals / np.sum(exp_vals, axis=1, keepdims=True)
    return probs, a1
    
```

```python
def calculate_loss(model, data_x, data_y):
    probs, _ = forward(model, data_x)
    probs[range(num_examples), data_y] -= 1
    probs = probs ** 2
    
    return probs.sum()/num_examples
```

```python
def predict(model, data_x):
    probs,_ = forward(model, data_x)
    
    return np.argmax(probs, axis=1)
```

```python
def build_model(data_x, data_y, nn_hdim=5, epochs = 1600, print_loss=False, epsilon=0.0001):
    W1 = np.random.randn(nn_input, nn_nodes) / np.sqrt(nn_nodes)
    b1 = np.zeros((1, nn_nodes))
    W2 = np.random.randn(nn_nodes, nn_output) / np.sqrt(nn_output)
    b2 = np.zeros((1, nn_output))

    model = { 'W1': W1, 'b1': b1, 'W2': W2, 'b2': b2}  
    
    for i in range(epochs):
        probs, a1 = forward(model, data_x)
        
        delta3 = probs
        delta3[range(num_examples), data_y] -= 1
        dW2 = (a1.T).dot(delta3)
        db2 = np.sum(delta3, axis=0, keepdims=True)
        delta2 = delta3.dot(W2.T) * (1 - a1**2)
        dW1 = np.dot(data_x.T, delta2)
        db1 = np.sum(delta2, axis=0)
 
        # Gradient descent parameter update
        W1 -= epsilon * dW1
        b1 -= epsilon * db1
        W2 -= epsilon * dW2
        b2 -= epsilon * db2
        
        if print_loss and i%100==0:
            print(f"Epoch {i}: Cost={calculate_loss(model, data_x, data_y)} Val_acc={round((predict(model, X_test) == y_test).sum()/len(y_test) * 100, 2)}")
        
        model = { 'W1': W1, 'b1': b1, 'W2': W2, 'b2': b2}
    return model
```

```python
m = build_model(X_train, y_train,10, print_loss=True)
```
